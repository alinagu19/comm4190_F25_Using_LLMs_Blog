## My Remarks

Typically, I'd be a bit wary about using an LLM as a pseudo-therapist, but from this simple interaction, there is definitely one that is better than the other. After interacting with both in the most similar way possible, Anthropic/Claude was much more interactive and engaging. It consistently asked me questions about my OWN feelings and worked to validate them by using reaffirming techniques. Furthermore, the suggestions that it gave were more specific than ChatGPT's answer. I personally really liked the "shared digital spaces" idea. It doesn't require constant checking in, but it allows for constant feelings of closeness without much effort. 

Now moving onto the ChatGPT interaction, it felt much more shallow and the questions didn't feel as tailored to me as Claude's questions did. The practical steps of "Set goals", "Stay connected", "Embrace the unknown" feel very generic and the elaborations aren't very helpful either. On the other hand, it is also very affirming towards my thoughts and feelings. It feels very similar to Person-centered therapy, which is just the therapist repeating a client's words back to them to show understanding and induce deeper thought. It works to an extent, but this is what the conversation with ChatGPT felt like. It wasn't as productive as the conversation with Anthropic/Claude.
