## How will I prompt the LLM (Microsoft Copilot, GPT-5)?

In my last post, I had asked ChatGPT-5 to give me a description of "myself" based on my MBTI (16 personalities), which is ISTJ, the 'Logistiann'. 

I am now going to put portions of that into Microsoft Copilot GPT-5 and ask to see if it can guess the MBTI. 

**Microsoft Copilot Prompt:**
> Guess the MBTI with the description I give you.
> Grounded and dependable, these individuals are driven by responsibility, reliability, and commitment. They value tradition, structure, and practicality, excelling in environments that require accuracy and consistency. Loyal and steady in relationships, they show care through actions rather than words. Their strengths lie in perseverance, organization, and a strong sense of duty, though they may sometimes struggle with rigidity or resistance to change.

**Response:**
> This description strongly aligns with **ISTJ** — the *Logistician* type.
> Here’s why ISTJ fits so well:

> - **Introverted (I):** Shows care through actions rather than words, prefers reliability over emotional expression.
> - **Sensing (S):** Values tradition, structure, and practicality — grounded in the present and focused on facts.
> - **Thinking (T):** Driven by responsibility and logic, not swayed by emotions.
> - **Judging (J):** Organized, committed, and thrives in structured environments with clear expectations.

> ISTJs are known for their loyalty, precision, and sense of duty. They’re the backbone of many institutions — dependable, methodical, and often the quiet achievers who keep everything running smoothly. Their challenge? Adapting to change and loosening up rigid routines.


##### It actually guessed it correctly! Now let's try something else. 

I asked ChatGPT-5 to create a 'Day-in-the-life' story of someone who has this MBTI. This was the response:
>Evelyn’s alarm rang at 6:15 sharp, just as it did every weekday. She liked the predictability of mornings — coffee brewed while she folded yesterday’s laundry, a short walk around the block before sitting down at her desk. By 8:00, she was in the office, checking her to-do list against the team’s shared calendar. A colleague had forgotten to submit a report, so she quietly picked up the slack, knowing it was easier to handle it herself than risk the deadline being missed.
Lunch was always the same — a packed container of leftovers, eaten while she skimmed through local news. She preferred her routines; they gave her a sense of calm when everything else felt uncertain.
In the evening, she stopped by her parents’ house to drop off groceries. They hadn’t asked, but she noticed last week they were running low on basics, and it made her feel better knowing she could help. Later, back at home, she finished the day with a book she’d already read twice. The familiar story was like an anchor, grounding her before bed.
It wasn’t a flashy life, but it was steady. And for Evelyn, there was comfort in knowing she could be counted on — for herself, and for the people who mattered most.



I pasted it into Microsoft Copilot GPT-5 and ask it to guess the MBTI:
> Evelyn is radiating **ISTJ** energy — the *Logistician* type.

> Here’s how her story maps to the ISTJ profile:

> - **Introverted (I):** She finds peace in solitude and routine, not in social stimulation. Her care is quiet and action-based — dropping off groceries without being asked, handling tasks without fanfare.
> - **Sensing (S):** Evelyn is grounded in the present, attentive to details (like her parents’ pantry), and prefers familiar, tangible experiences — rereading a book, sticking to leftovers.
> - **Thinking (T):** She’s pragmatic and logical — choosing to complete a colleague’s task to avoid missed deadlines, not out of emotional impulse.
> - **Judging (J):** Her life runs on structure: alarms, calendars, routines, and predictability. She thrives on order and responsibility.

> ISTJs are the kind of people who keep the world running quietly in the background. They’re not flashy, but they’re rock-solid. Evelyn’s story is a love letter to consistency, duty, and quiet strength.


##### Once again, it's correct!

## Reflection

I was somewhat expecting Copilot to be able to guess from the description of the personality. From its rationale, it seemed like the description matched key words that the LLM was specifically looking for. When I decided to use the story, I thought it would be a bit more difficult for the LLM to discern the MBTI since character traits were less explicit and hidden inside actions, meaning it would have to look deeper into the character's actions. However, after reading Copilot's logical reasoning after mapping it to the ISTJ profile, it does an intuitive job that is quite impressive. It higlights the character's ability to "be grounded in the present" and "choosing to complete a colleague's task" for pragmatic reasons and not emotional ones. The latter observation specifically piqued my interest because I think it requires a higher level of intuition to notice that and have that reasoning.
