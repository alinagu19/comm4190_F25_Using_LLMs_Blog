![](CNNs.jpg)

I am currently reading **A Brief History of Intelligence** by Max Bennet and I came across a section called, "The Problems of Pattern Recognition", which talks about types of learning and recognition from both a biological and computational standpoint. In particular, it addresses what Bennet dubs as "The Continual Learning Problem" and how computers recognize patterns. One thing that stood out to me when I was reading this chapter was the discussion about a type of network called "convolutional neural networks". I got really excited about this because I had just learned about this topic in my other class, BIOL 4536 - Computational Biology and Biological Modeling, and just had a homework assignment about it! (this homework assignment took way too long, so I thought I'd share and connect it to the book!)

Max Bennet explains it in fairly layman's terms. It started when a computer scientist, Kunihiko Fukushima wanted to get computers to recognize objects in pictures, because, after all, our brains can do it. For us, visual processing is hierarchical with different levels recognizing more simple or more complex features. Fukushima attempted to replicate that by designing a new architecture of artificial neural networks that differed from a fully connected neural network. Input photos would be decomposed into several feature maps, which contained information about the location of the feature within the original input picture. This output was then compressed and passed to another set of maps and so on and so forth. Verdict: IT WORKED

Still though, CNNs are still a poor approximation of how pattern recognition occurs in the brain. For example, even a fish brain, which is more simplistic than a mammalian brain, has the abilitiy to understand changes in rotation of 3D objects. That, for one, is something hat CNNs still struggle with. It would require incredible amounts of data to have that same understanding of orientation.
