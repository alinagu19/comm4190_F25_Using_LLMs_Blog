## *Breakthrough 5: Speaking and the First Humans*

Communication isn't some novel thing that started with humans. Organisms have been "communicating" with each other long before even the first animals came about. Single-celled organisms emit chemical signals to share information about the environment and fish even use electrical signals to attract mates. Chimps even use hunreds of different gestures and vocalizations to "speak" to each other. So what is so special about human communcation. Drum roll please...it's LANGUAGE!

Human language differs from other animal communication in two ways. First, **declarative labels** doesn't exist outside of human language. This essentially means assigning an object or behavior an arbitary symbol, the way we call the big green things outside trees or shuffling our feet back and forth walking. The second way is **grammar**. Obviously, not all human languages have the same usage of grammar, but it's still present and equally complex. Language not only allows us to communicate, but it also lets us peer into others' minds. This transferrence of thoughts allows us to construct common myths, which, believe it or not, is the backing factor behind society's coordinated behavior. Thus, this is how we pass down social etiquette, values, stories, and even morals. The chapter also included a really cool graph with millions of years on the x-axis and idea complexity on the y-axis. Essentially, idea complexity remains stagnant and flat until the emergence of language, which enables cumulative ideas. From there, idea complexity shoots up very steeply. 

Bennet then goes to talk about the emergence of the human hive mind, which is an interesting theory that explores how individuals connect and share intelligence. This ranges from the molecular scale, like neurons coordinating within our own brains to reach a neurological consensus) to collective human behavior, covering something like groupthink. He believes that first words may have emerged from proto-conversations between parents and children, ensuring surival. This argument would explain two things. First, it wouldn't require group selection (a proposed mechanism of evolution where natural selection works at the group level rather than the individual) and would work through common kin selection (a different evolutionary mechanism driven by the principle of individuals helping genetically related individuals survive). Second, apparently the learning program for language is most promiment in the interactions between parent and child. From there, the theory proposes the opportunity for using language between non-kin then became possible. The next topic of discussion is about gossip, which Robin Dunbar (an anthropologist) proposed is the most natural activity we use language for. It even has inherent rewards and punishmens. Spread of gossip enables more rewarding of altruistic behaviors because the spread of good deeds tends to lead to more social praise whereas negative behaviors are punished socially. The figure that Bennet includes explains this very well. It is essentially a circular cycle where "more gossip and punishment of violators" creates pressure for "more altruism towards non-kin", which creates pressure for "more language skills", and then that creates pressure for more gossip and punishment again.
