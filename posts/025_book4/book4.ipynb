## *Breakthrough 4: Mentalizing*

A lot of the discussion in this chapter revolved around primates. In particular, Chapter 17, "Moneky Hammers and Self-Driving Cars" stood out the most to me. It begins with a retelling of Jane Goodall's story when she followed a local tribe of chimpanzees in Gombe, Tanzania. She noticed two chimps taking branches, stripping the leaves off of them, and sticking them into termite mounds. They were using TOOLS. Although tool use was originally thought to be unique to humans, primates use them to quite a sophisticated degree as well. A "monkey mirror" experiment was also introduced in this chapter. Researchers had placed electrodes throughout a macaque monkey's brain to determine which areas of the brain responded to different types of hand movements. The researchers were eating their lunch when they noticed the electrodes connected to the monkey's brain were picking up signals. The monkey itself was not moving, but when the researchers moved their hands, the brain areas that were responsible for corresponding movements in monkeys also lit up. They discovered "mirror neurons"! 

Another interesting chapter in this section was titled "Why Rats Can't Go Grocery Shopping" (I quite like these chapter names) because it introduced a new term that I had never heard before, "frugivore", which is a fruit-based diet. It's intriguing how much cognition is required for this type of diet compared to carnivores or herbivores for example. Due to a tight window for ripeness and also competition with other frugivores, the animals on these diets (primarily primates) really have to plan ahead and also track the population of fruit. This whole portion of the chapter then segues into "anticipation", which ,right now in the book's timeline of intellectual development, is a foreign concept. 

It's really something that we take for granted, but anticipation is quite an advanced mechanic for our neocortex. Bennet then provides a bit of a long-winded tangent to come to this point: it is the foundation for our ability to make long-term plans. I wanted to note that this last point is cool and all, but I can't help but wonder...how does this connect to the bigger picture?

Bennet does a really good job of contextualizing these breakthroughs and detailing how they came about, but, after the chapter, I am still left wondering "now how are they going to be able to get AI to be able to do that?". Looking back at the rest of the book, this curiosity resurfaces quite often. I think we can all agree that the way intelligence has developed from the first fungi to early mammals to primates to humans is rather astounding. However, if the goal is to provide the same intelligence to AI, how is that going to be done practically?
