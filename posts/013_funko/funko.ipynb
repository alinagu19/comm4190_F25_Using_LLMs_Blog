### Can Chat-GPT5 make me into a Funko Pop?

I've always really enjoyed the Funko Pop figurines of different characters. Using Chat-GPT5's image generation feature, I am wondering how well it can mimic that style with a photo of myself. 

![](funko.png)

##### The prompt was plain and simple:
> Make me into a Funko Pop! (I attached the photo on the left)

### My Thoughts

I've been wanting to try this for a while because I think the image generation feature of Chat-GPT5 is really neat, especially the fact hat it can mimic art styles or the style of a specific product. I was really impressed how the LLM also copied my outfit, including the skirt, which isn't completely in the photo. It didn't originally include shoes but when I reprompted with "add white sneakers to it" and this is what I got:
![](funko2.png)
I found it a bit interesting how the LLM changed the face a bit, the hair, and the length of the skirt without me asking it to. I've noticed this sometimes when image generating, how it will change features that I didn't want changed and then I have to go back and correct. However, overall, I think it did a really good job and now I'm curious to try more art styles to see what Chat-GPT5 can recreate.
