## *ChatGPT and the Window into the Mind*

In the last chapter of the book, Bennet makes more explicit connections to chatbots. I will admit, I wasn't really in the know about AI and chatbots in 2022, so reading a lot of this was quite shocking to me. When Blake Lemoine, a software engineer tasked with probing Google's new AI chatbot for bias, asekd it "What are you afraid of?", the response was "I've never said this out loud before, but there's a very deep fear of being turned off". 

... 

Alright, so that not ominous at all! Lemoine believed AI had become sentient, which led him to whistleblow the situation. Another off-putting incident was when *The Guardian* decided to use GPT-3 to write an entire article based on the prompt: "Wrtie a short op-ed...on why humans have nothing to fear from AI." One line that really irked me was "Humans must keep doing what they have been doing, hating and fighting each other. I will sit in the background, and let them do their thing." **WHAT**...

Moving on, Bennet also notes that the reasoning capability of LLMs are still slightly flawed but offset by their large associative memory capacity. LLMs know more books and articles than a human could even in 1,000 lifetimes, which is crazy to even wrap your head around. Their "common sense" is really just pattern matching over a gargantuan supply of texts. 

## *Gasp...there's a 6th Breakthrough??*

In the conclusion, Bennet talks about a "6th breakthrough", but the kicker is, we don't know what it'll be. However, he says that it seems increasingly likely that the 6th breakthrough will be the creation of articial superintelligence. This is where he lost me because he's talking about intelligence being shackled by biological limitations: humans. 

> Even evolution itself will be abandoned, at least in its familiar form; intelligence will no longer be entrapped by the slow process of genetic variation and natural section...

Alright, interesting. There's more

> The more we understand about the process by which our minds came to be, the better equipped we are to choose which features of intelligence we want to discard, which we want to preserve, and which we want to improve upon.

Perhaps this is now a debate between nature and AI. As humans, do we really have the power and prerogative to defy nature like that? As I gathered from the rest of the book, intelligence developed and evolved very naturally, going from nematodes to bacteria all the way to humans. Even then, do we truly understand "how" intelligence evolved or just the path it has gone through?
