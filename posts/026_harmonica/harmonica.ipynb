## **Teaching Breathing**

### ASSISTANT

### üí® Step 1: Understand the Problem

Most beginners get out of breath for two main reasons:

1. **Breathing with your chest**, not your diaphragm.
2. **Controlling the airflow improperly**: using too much force.
3. **Exhaling too much while blowing**, and not enough air intake while drawing. It creates an imbalance.

---

### üßò Step 2: Diaphragmatic (Belly) Breathing

Instead of breathing with your chest, we want to breathe from your diaphragm‚Äîthis is called **belly breathing**.

üîπ **Try this without the harmonica first**:

1. Sit or stand up straight.
2. Put one hand on your stomach, one on your chest.
3. Breathe in slowly through your nose. Try to only move the hand on your belly.
4. Exhale slowly through your mouth.
5. Repeat several times.

This trains your body to take deeper, more relaxed breaths from the belly, not shallow chest breaths.

üëâ Practice for 2 minutes a day. Most harmonica pros use this breathing method.

### üéØ Step 3: Control Airflow ‚Äì "Gentle Breaths"

You don‚Äôt need to blow or suck hard to get a sound from the harmonica. Use a **gentle, steady breath**‚Äîthe same kind of breath you‚Äôd take to quietly breathe on a window to fog it up.

‚úÖ Your goal is a relaxed, clear tone‚Äîno force!
