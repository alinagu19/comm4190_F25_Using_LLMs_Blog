### Quick styling tips for graduation

* Shoes: low block heels or classy mules — comfortable for walking across a stage and photogenic.
* Outer layer: a lightweight blazer or cropped cardigan if your ceremony’s in an air-conditioned hall.
* Jewelry: simple studs + a delicate necklace keeps the look refined; pick one statement piece (earrings or bracelet) rather than many.
* Bag: a small clutch or mini shoulder bag to hold phone/ID — keep it neutral or metallic.
* Undergarments: consider strapless/adhesive options for corsets or cowl backs, and seamless nude undergarments for satin.
