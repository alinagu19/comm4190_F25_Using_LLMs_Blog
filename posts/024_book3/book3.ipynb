### *Breakthrough 3 - Simulating*

This portion of the book focused on the third breakthrough of intelligence, simulating. It went into depth on the neocortex, which was the primary new brain strucutre that emerged in early mammals. The notion of "simulating" that came along with this new evolutionary breakthrough was the idea of learning and advanced memory. One really substantial type of learning that emerged is known to us as "trial and error", which if you truly think about it, is still the crux of how we learn even to this day. As humans, "you live and you learn" is a massively common saying and for good reason too. Trial and error is probably the most fundamental type of learning and as basic as it is, it works. From there, the agranular prefrontal cortex (aPFC) enabled early mammals to engage in vicarious trial and error, combining simulation with basic trial and error. Early mammals were then able to simulate future actions and imagine outcomes to decide which path to take. 

Counterfactual learning built on this further by offering a better solution to the credit assignment problem. The credit assignment problem is a challenge for learning systems to decide which specific actions are responsible for a final outcome. 

I was curious in how this all ties back to AI, and after doing some reading, this really is all the basis of machine learning. In terms of counterfactual learning, AI (or whatever system it is) draws from alternative realities/hypothetical situations to past events to learn and make better choices. Similarly, figuring out solutions to the credit assignment problem enables learning systems to deduce how much weight a specific path has towards a certain outcome, commonly a reward. Since feedback can usually be delayed (coming after a long sequence of decisions), it's difficult for the system to decide which decision is actually responsible for the reward. Solving the credit assignment problem is crucial for this.
